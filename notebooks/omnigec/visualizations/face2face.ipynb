```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Sample df_long creation for demo purposes
# Assuming df_long already exists
# agg = (df_long.groupby(['track','language','model'])['gleu'].mean().reset_index())

# Use the actual agg from before
agg = (
    df
    .groupby(['track', 'language', 'model'])['gleu']
    .mean()
    .reset_index()
)

# Prepare RdYlGn colormap with gray for NaNs
cmap = plt.get_cmap("RdYlGn").copy()
cmap.set_bad("lightgray")

for track in agg['track'].unique():
    sub = agg[agg['track'] == track]
    mat = sub.pivot(index='language', columns='model', values='gleu').astype(float)
    data = np.ma.masked_invalid(mat.values)

    models = mat.columns.tolist()
    # Create two-line labels
    labels = []
    for m in models:
        if "(" in m and m.endswith(")"):
            name, sep, rest = m.partition(" (")
            labels.append(f"{name}\n({rest}")
        else:
            labels.append(m)

    # Boundary index
    idx_llama = models.index("LLaMA-3-8B (MultiGEC)")
    boundary = idx_llama - 0.5

    height = 8 if track == 'minimal' else 6
    fig, ax = plt.subplots(figsize=(12, height))
    im = ax.imshow(data, aspect='auto', cmap=cmap, alpha=0.5)
    ax.axvline(boundary, color='black', linewidth=1)
    ax.grid(False)

    for side in ["top", "bottom", "left", "right"]:
        ax.spines[side].set_visible(True)    # make sure itâ€™s shown
        ax.spines[side].set_linewidth(1)   # thickness (adjust to taste)
        ax.spines[side].set_color("black")

    ax.set_ylabel('Language')
    ax.set_xlabel('Model')
    ax.set_yticks(range(len(mat.index)))
    ax.set_yticklabels(mat.index)
    ax.set_xticks(range(len(models)))
    ax.set_xticklabels(labels, rotation=45, ha='right')

    # Annotate
    row_max = np.nanmax(mat.values, axis=1)
    for i in range(mat.shape[0]):
        for j in range(mat.shape[1]):
            val = mat.iat[i,j]
            txt = "â€“" if np.isnan(val) else f"{val:.2f}"
            weight = 'bold' if not np.isnan(val) and np.isclose(val, row_max[i]) else 'normal'
            ax.text(j, i, txt, ha='center', va='center', fontweight=weight, color='black')

    # Group headers under axis
    total = len(models)
    frac_our = idx_llama / total
    frac_comp = 1 - frac_our
    ax.text(frac_our/2, 1.1, 'Our Results', transform=ax.transAxes, ha='center', va='top', weight='bold')
    ax.text(frac_our + frac_comp/2, 1.1, 'MultiGEC-25 Comp.', transform=ax.transAxes, ha='center', va='top', weight='bold')

    plt.colorbar(im, ax=ax, label='GLEU')
    plt.tight_layout()
    plt.savefig(f"{track}.png", dpi=300)
    plt.show()
```
